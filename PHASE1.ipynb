#List vs Numpy array

```python
import time

py_list = [1,2,3]
print("Python list multiplication is: ", py_list*2)

np_array = np.array([1,2,3]) #element-wise multiplication
print("Python array multiplication is : ", np_array*2)

start = time.time()
py_list = [i* 2 for i in range(1000000)]
end = time.time()
total_time = end - start
print("pylist :",total_time)

start = time.time()
np_array = np.arange(100000) * 2
end = time.time()
total_time = end - start
print("pyarray :",total_time)
```

creating array from scratch

```python
zeros = np.zeros((3,4))
print("zeros array :\n",zeros)

ones = np.ones((3,3))
print("ones array :\n",ones)

full = np.full((2,2),7)
print("full array :\n",full)

random = np.random.random((2,3))
print("random array :\n",random)

seq = np.arange(0,10,2)
print("sequence array :\n",seq)
```

Vector, Matrix and Tensor


```python
vector = np.array([1,2,3])
print("Vector :",vector)

matrix = np.array([[1,2,3],
                   [4,5,6]])
print("Matrix :", matrix)

tensor = np.array([[[1,2],[3,4]],
                   [[5,6],[7,8]]])
print("Tensor :", tensor)
```

Array Propeties

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape:",arr.shape)
print("dimension :",arr.ndim)
print("size:",arr.size)
print("Dtype :",arr.dtype)
```

Array Reshaping

```python
arr = np.arange(12)
print("Original array:",arr)

reshaped = arr.reshape(3,4)
print("\nReshaped array:",reshaped)

flattened = reshaped.flatten()
print("\nFlattened array:",flattened)

# ravel (returns view,instead of copy)
raveled = reshaped.ravel()
print("\nRaveled array:",raveled)

# Transpose
transpose = reshaped.T
print("\nTransposed array:",transpose)
```
