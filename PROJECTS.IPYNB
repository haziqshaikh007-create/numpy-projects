PROJECT1 - Students Marks Analayzer

```python
###[Eng,Urdu,Math]
marks = np.array([
[78,85,90],
[65,70,60],
[88,92,95],
[50,55,58],
[90,91,89]
])

print("Average Marks Per student:", np.mean(marks,axis=1))
print("Highest Scorer Per subject:", np.max(marks,axis=0))
print("Subject-wise Average:", np.mean(marks,axis=0))
student_avg = np.average(marks,axis=1)
print("Student who passed:", np.where(student_avg>60)[0])
```

Project2 - Sales Data Analyzer

```python
#product = day1,day2,day3,day4 
sales = np.array([
[200,300,250,400],         
[150,250,300,350],         
[400,450,500,550]         
])

#questions
print("Total Sales per day:", np.sum(sales,axis=0))
print("Total Sales per Product:", np.sum(sales,axis=1))
total_sell = np.sum(sales,axis=1)
print("Best selling product:",np.argmax(total_sell))
print("Growth per day:", np.cumsum(sales,axis=0))

#extras

print("Average sales per product:", np.mean(sales,axis=1))
total_sell = np.sum(sales,axis=0)
print("Lowest selling day:",np.argmin(total_sell))
growth = sales[:, -1] - sales[:,0]
print("Growth per product:", np.argmax(growth))
```

Project4- Random Dataset Analyzer

```python

data = np.random.randint(1,100,(10,5))
print(data)

#questions

print("Mean of data-set:", np.mean(data))
print("Median of data-set:", np.median(data))
print("Standard Deviation of data-set:", np.std(data))
print("Highest value of data-set:",np.max(data))
print("Lowest value of data-set:",np.min(data))
print("All values over 70:", data > 70)
print("Sorted data:",np.sort(data))
row_avg = np.mean(data,axis=1)
print("Row avg",row_avg,"Row with Highest average is :",np.argmax(row_avg))
col_avg = np.mean(data,axis=0)
print("Column avg",col_avg,"Column with Lowest average is :", np.argmin(col_avg))
print("Total sum:", np.sum(data))
```

Project5 - Recommendation System Logic

```python
ratings = np.array([
[5,4,0,1],
[4,0,0,1],
[1,1,0,5],
[1,0,0,4]
])

#Questions
avg_rating = np.mean(ratings,axis=0)
print("Average Rating per movie:", avg_rating)
print("Highest Rated movie:", np.argmax(avg_rating))
print("Lowest Rated movie:", np.argmin(avg_rating))
print("Movies not watched:", np.where(ratings == 0)[0])

for i in range(len(ratings)):
         unseen = np.where(ratings[i] == 0)[0]
         print(f"User {i+1} has not watched {unseen}")

user1_unseen = np.where(ratings[1] == 0)[0]
recommended = user1_unseen[np.argmax(avg_rating[user1_unseen])]
print("The recommended movie is:",recommended)

activity = np.count_nonzero(ratings, axis=1)
print("Most active user:", np.argmax(activity))

most_viewed = np.count_nonzero(ratings, axis=0)
print("The Most Times Rated is:", np.argmax(most_viewed))

user_avg_rating = np.mean(ratings,axis=1)
print(user_avg_rating,"User with highest average rating is:", np.argmax(user_avg_rating))

popularity = avg_rating * most_viewed
print(popularity,"Most popular movie is :", np.argmax(popularity))
```
