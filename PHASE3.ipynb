```python
import numpy as np
import matplotlib.pyplot as plt
```

Advance Operations with Business examples

```python
#  Data structure: [resturantid,2021,2022,2023,2024]

sales_data = np.array([
[1, 150000, 180000, 220000, 250000], #Paradise Biryani
[2, 120000, 140000, 160000, 190000], #Beijing Bites
[3, 200000, 230000, 260000, 300000], #Pizza Hub
[4, 180000, 210000, 240000, 270000], #Burger Point
[5, 160000, 185000, 205000, 230000] #Chai Point        
])

print("=== Zomato sales analysis ====")
print("\nSales data shape",sales_data.shape)
print("\n Sample data for first 3 resturants:", sales_data[:3])
```

```python
###total sales per year

print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:,1:],axis =0)
print("Total per each year",yearly_total)

#Minimum sale sper resturant

min_sales = np.min(sales_data[:,1:],axis=1)
print("Minimum sales of each resturant",min_sales)

max_sales= np.max(sales_data[:,1:],axis = 0)
print("Max sales _per_year ",max_sales)

avg_sales = np.average(sales_data[:,1:],axis=1)
print("Average sale per resturant :", avg_sales)
```

```python
cum_sum = np.cumsum(sales_data[:,1:],axis=1)
print("Total sum ", cum_sum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cum_sum,axis = 0))
plt.title("Average cumulative sale sacross all resturants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition :", vector1 + vector2)

print("\nVector multiplication:", vector1 * vector2)

print("\nDot product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1,vector2) /( np.linalg.norm(vector1) * np.linalg.norm(vector2) ))
print(angle)
```
